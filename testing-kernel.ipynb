```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import io, color
```

```python
img = io.imread("/kaggle/input/cnn-data-sources/yo.png")
```

```python
img
```

```python
plt.imshow(img)
```

```python
#Always is necesary only if its possible, transform the image to a gray scale
img_gray = color.rgb2gray(img)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray)
```

```python
#To detect borders need to use this values of filter
kernel = np.array([[-1, 0, 1],
                  [-1, 0, 1],
                  [-1, 0, 1]])
```

```python
kernel
```

```python
img_new = nd.convolve(img_gray, kernel)
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 10))
axes[0].imshow(img_gray, cmap = plt.cm.gray)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(img_new, cmap = plt.cm.gray)
axes[1].set_title("Convolutional")
axes[1].axis("off")
```

```python
kernel = np.array([[0, 0, 0],
                  [0, 1, 0],
                  [0, 0, 0]])
```

```python
img_new = nd.convolve(img_gray, kernel)
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 10))
axes[0].imshow(img_gray, cmap = plt.cm.gray)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(img_new, cmap = plt.cm.gray)
axes[1].set_title("example")
axes[1].axis("off")
```

```python
kernel = np.array([[1, 0, -1],
                  [1, 0, -1],
                  [1, 0, -1]])
```

```python
img_new = nd.convolve(img_gray, kernel)
```

```python
fig, axes = plt.subplots(1,2, figsize = (15, 10))
axes[0].imshow(img_gray, cmap = plt.cm.gray)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(img_new, cmap = plt.cm.gray)
axes[1].set_title("Test")
axes[1].axis("off")
```
