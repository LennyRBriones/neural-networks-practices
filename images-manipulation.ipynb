```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

## We used this data
# https://www.kaggle.com/datasets/alarcon7a/cnn-data-sources
```

```python
im = io.imread("../input/cnn-data-sources/Mara.jpeg")
```

```python
im
```

```python
im.shape
```

```python
plt.imshow(im)
```

```python
#Knowing that the color images is representated y layers, we can separete the layers

r = im[:,:,0]
g = im[:,:,1]
b = im[:,:,2]
```

```python
plt.imshow(r,cmap="gray")
#When we only call to 1 layer, the layer is represented in gray scale
```

```python
#Also we can make the transpose of the matrix
plt.imshow(r.T, cmap="gray")
```

```python
r.shape
```

```python
aux_dim = np.zeros([800,600])
#We make the auxiliar with the same shape
```

```python
#Now we need to stack the auxiliar and the red layer
np.dstack((r,aux_dim, aux_dim)).astype(np.uint8)
## The stack help us to make 3 dimentions, getting the possibility to show the red layer as red
## putting 0 in the rest of the layers
```

```python
red = np.dstack((r,aux_dim, aux_dim)).astype(np.uint8)
```

```python
plt.imshow(red)
```

```python
#making the green & blue layers
green = np.dstack((aux_dim, g, aux_dim)).astype(np.uint8)
blue = np.dstack((aux_dim, aux_dim, b)).astype(np.uint8)
```

```python
plt.imshow(blue)
```

```python
plt.imshow(green)
```

```python
## We can print the negative values substracting the 255 to the images values
im_neg_pos = 255 - im
plt.imshow(im_neg_pos)
```

```python
#Downgrade to the resolution

im32 = (im//32) * 32
plt.imshow(im32)
```

```python
im128 = (im//128) * 128
plt.imshow(im128)
```

```python
#Moving the matrix we can make a zoom or only show what we want
plt.imshow(im[270:530, 150:400])
```
